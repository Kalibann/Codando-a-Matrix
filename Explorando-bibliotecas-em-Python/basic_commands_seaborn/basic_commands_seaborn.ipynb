#### Importação

```python
import seaborn as sns
import pandas as pd
```

```python
data = pd.read_csv("heart.csv")
data.head()
```

#### Gráfico de Dispersão

```python
sns.scatterplot(data = data, x = "age", y = "chol")
```

#### Gráfico de Linha

```python
sns.lineplot(data = data, x = "age", y = "chol")
```

#### Gráfico de Barras

```python
sns.barplot(data = data, x = "thall", y = "chol")
```

#### Gráfico com Regressão

```python
sns.lmplot(data = data, x = "age", y = "thalachh")
```

#### Gráfico de Caixa

```python
sns.boxplot(data = data, x = "thall", y = "thalachh")
```

#### Gráfico de Correlação

```python
# Primeiramente, vamos calcular a correlação das colunas com o método corr()
corr = data.corr()

# Após isso, plotamos o gráfico
sns.heatmap(corr)
```
